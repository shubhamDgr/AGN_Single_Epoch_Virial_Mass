```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
data = pd.read_csv("E:/Projects/AGN_mass_project/data/shaw_2012.csv")
data = data.loc[:]
```

```python
x= data['WHb']
y= data['WMgII']
df = pd.DataFrame({'x': x, 'y': y})
```

```python
df['x'] = pd.to_numeric(df['x'], errors='coerce')
df['y'] = pd.to_numeric(df['y'], errors='coerce')
df = df.dropna(subset=['x', 'y'])
# print(df)

df['x'] = pd.to_numeric(np.log10(df['x']), errors='coerce')
df['y'] = pd.to_numeric(np.log10(df['y']), errors='coerce')
```

```python
result_og = stats.linregress(x=df['x'], y=df['y'])
result_og
```

```python
plt.scatter(df['x'],df['y'],  color= 'black')
plt.xlabel('log [FWHM (HÎ²)]')
plt.ylabel('log [FWHM (MgII)]')
plt.title('Shaw et al')
plt.plot(df['x'], df['x']*result_og.slope + result_og.intercept, color='red')
plt.show()
```
